```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

mu = 80
sigma = 7
x = np.random.normal(mu, sigma, size=500)


fig, ax = plt.subplots()


ax.hist(x, 20)
ax.set_title('Historgram')
ax.set_xlabel('bin range')
ax.set_ylabel('frequency')


fig.tight_layout()
plt.savefig("example_histogram.png")
plt.show()
```
